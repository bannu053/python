```python
import random;
```

```python
class Bank:
  Name= "HDFC"
  IFSC ="HDFC00678"
  Branch="Hyderabad"
  Accounts = []
  def __init__(self,Account_name,Account_type,Account_no,Balance) :
    self.Account_name=Account_name
    self.Account_type=Account_type
    self.Account_no=Account_no
    self.Balance=Balance
  def create_account():
    a=input("Enter name:")
    b=input("Enter account type:")
    c=random.randint(0, 100)
    print("Welcome to HDFC Bank!")
    print(f"Note your account number:{c}")
    customer=Bank(a,b,c,500)
    Bank.Accounts.append(customer)
  def deposit():
    d =int(input("Enter your account number:"))
    e= int(input("Enter amount:"))
    for i in Bank.Accounts:
      if i.Account_no==d:
        f=i.Balance
        break
    i.Balance += e
    print("Deposit Successfull")
    print(f"your balance:{i.Balance}")
  def withdraw():
    f =int(input("Enter your account number:"))
    g= int(input("Enter amount:"))
    for i in Bank.Accounts:
      if i.Account_no==f:
       break
    if i.Balance >=g:
      i.Balance -= g
    else:
      print("Insufficient Fund")
    print(f"your balance:{i.Balance}")
  def check_balance():
    f =int(input("Enter your account number:"))
    for i in Bank.Accounts:
      if i.Account_no==f:
       break
    print(f"your balance:{i.Balance}")





while True:
    print("""
1. Create Account
2. Deposit
3. Withdraw
4. Check Balance
5. Exit
""")

    choice = int(input("Enter choice: "))

    match choice:
        case 1:
            Bank.create_account()
        case 2:
            Bank.deposit()
        case 3:
            Bank.withdraw()
        case 4:
            Bank.check_balance()
        case 5:
            print(f"Thank you for choosing {Bank.Name} Bank, {Bank.Branch}!")
            break
        case _:
            print("Invalid choice")

```
